**Visualization of Result Output**

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="results_50_part_time.csv"
data=pd.read_csv(path)
data
```

```python
# Filter rows where 'two_step' is True
data_true = data[data['two_step'] == True]

# Sort by threshold for proper line plotting
data_true = data_true.sort_values('threshold')

# Plot each metric
plt.figure(figsize=(10, 6))
plt.plot(data_true['threshold'], data_true['recall'], label='Recall', marker='o')
plt.plot(data_true['threshold'], data_true['precision'], label='Precision', marker='s')
plt.plot(data_true['threshold'], data_true['f1_score'], label='F1-Score', marker='^')

# Plot settings
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Accuracy, Precision, F1-Score vs Threshold for Two-Step Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter rows where 'two_step' is True
data_false = data[data['two_step'] == False]

# Sort by threshold for proper line plotting
data_false = data_false.sort_values('threshold')

data_false

```

```python
# Plot execution time vs threshold for both subsets
plt.figure(figsize=(10, 6))

plt.plot(data_true['threshold'], data_true['execution_time'], label='Two Step = True', marker='o', color='blue')
plt.plot(data_false['threshold'], data_false['execution_time'], label='Two Step = False', marker='s', color='orange')

# Plot settings
plt.xlabel('Threshold')
plt.ylabel('Execution Time (s)')  # Adjust unit if needed
plt.title('Execution Time vs Threshold')
plt.legend()
plt.grid(True)
plt.show()
```
